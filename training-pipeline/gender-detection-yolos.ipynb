```python
from ultralytics import YOLO
import torch
```

```python
# Load YOLOv8 model
model = YOLO("yolov8n.pt")  # You can replace with your model or a pretrained one
```

```python
dataset_path = "/kaggle/input/yolo-gender-data/men-women-dataset/data.yaml"  # Adjust accordingly
```

```python
# Set optimizer directly in the YOLO model call
train_results = model.train(
    data = dataset_path,  # Path to dataset YAML
    epochs = 120,
    imgsz = 640,
    device = "cuda" if torch.cuda.is_available() else "cpu",
    batch = 16,

)


# After training, you can view the results
print("Training complete!")

# Optionally, you can export the best model checkpoint after training
best_model_path = "best.pt"  # This is the default filename for the best checkpoint saved
print(f"Best model saved at: {best_model_path}")
```
